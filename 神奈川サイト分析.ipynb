# Stations


```python
! wget https://www.pref.kanagawa.jp/sys/taikikanshi/kanshi/data/V501Station.json

with open("V501Station.json") as f:
    stations = json.load(f)
stations
```

```python
! wget https://www.pref.kanagawa.jp/sys/taikikanshi/kanshi/data/V503StationItem.json

with open("V503StationItem.json") as f:
    station_items = json.load(f)
station_items
```

```python
! wget  https://www.pref.kanagawa.jp/sys/taikikanshi/kanshi/data/V502Item.json

with open("V502Item.json") as f:
    items = json.load(f)
items
```

```python
# 横軸に必要なitemをtableから抽出する。
columns = set()
for scode, sinfo in table.items():
    columns |= set(sinfo.keys())
columns
```

```python
icodes = sorted(list(columns))
icodes
```

```python
# header row
header = ["測定局", "区分"]
for icode in icodes:
    if icode in items:
        iinfo = items[icode]
        header.append(f'{iinfo["itemName"]}, {iinfo["unit"]}')
    else:
        header.append(icode)
header
```

```python
body = [header]
for scode, sinfo in table.items():
    if scode in stations:
        row = [stations[scode]["name"], stations[scode]["cate"]]
        for icode in icodes:
            if icode in sinfo:
                row.append(sinfo[icode])
            else:
                row.append("---")
    else:
        row = [scode]
    body.append(row)

body
```
